- **MANIPULAÇÃO DE DADOS**

```python
coletardados_prices = [x for x in dados["Price"]]
```

```python
media_dos_preços = []
cont = 0 
for y in range(len(coletardados_prices)):
    cont += 1
    mani = coletardados_prices[y].replace('Rs.','',).replace('Lakh','').strip()
    mani = mani.split(' - ')
    media_dos_preços.append(mani)
print(media_dos_preços)
```

```python
extrair_media = []
cont = 0
for x in range(len(media_dos_preços)):
    cont += 1
    if media_dos_preços[x] == 2:
        media = (float(media_dos_preços[x][0]) + float(media_dos_preços[x][1]))
    else:
        media = (float(media_dos_preços[x][0]))
    media = round(media * 1000000)
    extrair_media.append(media)
```

- **FILTRAGEM E SELEÇÃO**

```python
dados = dados.fillna(0)
```

```python
filtrar = dados['Seating.Capacity']
```

```python
filtrar.head()
```

```python
extrair_numero = [str(filtrar[x]).replace('Seater', ' ').replace(',', ' &') for x in range(len(filtrar))]
extrair_numero_1 = []
for e in range(len(extrair_numero)):
    manipu_la = extrair_numero[e].strip().split('&')
    extrair_numero_1.append(manipu_la)
```

```python
extrair_numero_2 = []
for i in range(len(extrair_numero_1)):
    if len(extrair_numero_1[i]) == 3:
        media = round(int(extrair_numero_1[i][0]) + int(extrair_numero_1[i][1]) + int(extrair_numero_1[i][2])/3)
        extrair_numero_2.append(media)
    elif len(extrair_numero_1[i]) == 2:
        media = round(int(extrair_numero_1[i][0]) + int(extrair_numero_1[i][1])/2)
        extrair_numero_2.append(media)
    else:
        media = round(int(extrair_numero_1[i][0]))
        extrair_numero_2.append(media)
```

```python
dados['Capacidade'] = pd.Series(extrair_numero_2)
```

```python
filtroeselecao_df = dados.query('Capacidade > 5')
```

```python
dados.columns = ['Nome', 'Preço', 'Quilometragem', 'Motor', 'Segurança', 'Tipo_de_Combustível', 'Transmissão', 'Capacidade de Assentos', 'Capacidade']
```

- **AGREGAÇÃO DE DADOS**

```python
manipulação_quilometragem = [ x for x in dados["Quilometragem"]]
manipulação_quilometragem_1 = [str(manipulação_quilometragem[r]).replace('kmpl', '') for r in range(len(manipulação_quilometragem))]
manipulação_quilometragem_2 = [str(manipulação_quilometragem_1[r]).split('to') for r in range(len(manipulação_quilometragem_1))]
manipulação_quilometragem_3 = [((float(manipulação_quilometragem_2[r][0]) + float(manipulação_quilometragem_2[r][1])) / 2) if len(manipulação_quilometragem_2[r]) == 2 else float(manipulação_quilometragem_2[r][0]) for r in range(len(manipulação_quilometragem_1))]
manipulação_quilometragem_4 = [round(manipulação_quilometragem_3[x],2) for x in range(len(manipulação_quilometragem_3))]
```

```python
dados["Quilometragem_Media"] = pd.Series(manipulação_quilometragem_4)
```
